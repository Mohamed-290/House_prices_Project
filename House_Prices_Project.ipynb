### make the function to convert the data by your self                           ***very important***

```python
def convert (condition):
  if condition == 'Poor':
    return 1
  elif condition == 'Fair':
    return 2
  elif condition == 'Average':
    return 3
  elif condition == 'Good':
    return 4
  elif condition == 'Very Good':
    return 5
```

```python
data['condition']= data['condition'].apply(convert)
```

# Polynomial_Regression Algorthism

```python
import numpy as np
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split
```

```python
input_model_train, input_model_test, output_model_train, output_model_test = train_test_split(input_model, output_model, test_size=0.20)
degree = 2
polyreg = make_pipeline(PolynomialFeatures(degree))
polyreg.fit(input_model_train, output_model_train)
```

```python
score1 = polyreg.score(input_model_train, output_model_train)
score2 = polyreg.score(input_model_test, output_model_test)
print("Train_score: ", score1)
print("Test_score: ", score2)
```
